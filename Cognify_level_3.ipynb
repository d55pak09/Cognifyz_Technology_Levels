#### Analyze the text reviews to identify the most common positive and negative keywords

```python
# showing the unique values of columns
df['Rating text'].unique()
# counting the values
df['Rating text'].value_counts()
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Load NLTK stopwords
nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english'))

# Tokenize and process the reviews
positive_keywords = []
negative_keywords = []

for review in df['Rating text']:
    tokens = word_tokenize(review.lower())
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    if 'Good' in tokens or 'Average' in tokens or 'Excellent' in tokens or 'Very Good' in tokens:  # Add more positive keywords
        positive_keywords.extend(tokens)
    if 'Poor' in tokens:  # Add more negative keywords
        negative_keywords.extend(tokens)

# Calculate word frequencies
positive_freq_dist = FreqDist(positive_keywords)
negative_freq_dist = FreqDist(negative_keywords)

# Get most common positive and negative keywords
most_common_positive = positive_freq_dist.most_common(10)
most_common_negative = negative_freq_dist.most_common(10)

# Print the results
print("Most common positive keywords:")
print(most_common_positive)

print("\nMost common negative keywords:")
print(most_common_negative)
```

#### Calculate the average length of reviews and explore if there is a relationship between review length and rating.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate review lengths
data['Review Length'] = data['Rating text'].str.len()

# Calculate average review length
average_review_length = data['Review Length'].mean()

# Explore relationship between review length and rating
plt.figure(figsize=(8, 4))
plt.scatter(data['Rating text'], data['Review Length'], alpha=0.5)
plt.title('Relationship between Review Length and Rating text')
plt.xlabel('Rating text')
plt.ylabel('Review Length')
plt.show()

# Print the average review length
print("Average Review Length:", average_review_length)
```

#### Identify the restaurants with the highest and lowest number of votes.

#### Determine if higher-priced restaurants are more likely to offer these services.

```python
# Map price range categories to numerical values
price_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
data['Price range'] = data['Price range'].map(price_mapping)

# Categorize restaurants by price range
def categorize_price(price):
    if price < 2:
        return 'Low'
    elif price < 3:
        return 'Medium'
    else:
        return 'High'

data['Price Range'] = data['Price range'].apply(categorize_price)

# Convert categorical columns to numerical using binary encoding
services = ['Has Online delivery', 'Has Table booking']
data_encoded = pd.get_dummies(data, columns=services, drop_first=True)

# Calculate service occurrence by price range
price_range_services = data_encoded.groupby('Price Range').sum()

# Plot a heatmap to visualize service occurrence by price range
plt.figure(figsize=(10, 6))
sns.heatmap(price_range_services, annot=True, cmap='YlGnBu')
plt.title('Service Occurrence by Price Range')
plt.xlabel('Services')
plt.ylabel('Price Range')
plt.show()
```
