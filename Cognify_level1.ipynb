#### Determine the top three most common cuisines in the dataset.

```python
#1 Determine the top three most common cuisines in the dataset.
cuisines_count = data['Cuisines'].value_counts()

# top 3 cuisines
top_cuisines = cuisines_count.head(4)
top_cuisines.to_frame()
```

#### Calculate the percentage of restaurants that serve each of the top cuisines.

```python
#2 Calculate the percentage of restaurants that serve each of the top cuisines.

# Calculate the total number of restaurants
total_restaurants = len(data['Restaurant Name'])

# Calculate the percentage of restaurants for each cuisine
cuisine_percentages = (top_cuisines / total_restaurants) * 100

cuisine_percentages.to_frame()
```

#### Identify the city with the highest number of restaurants in the dataset

```python
# Use value_counts() to count the number of restaurants in each city
restaurant_counts = data['City'].value_counts()

# Get the city with the highest number of restaurants
city_with_highest_restaurants = restaurant_counts.idxmax()

# Get the count of restaurants in the city with the highest number of restaurants
highest_restaurants_count = restaurant_counts.max()
print(f"The city with the highest number of restaurants is: {city_with_highest_restaurants}")
print(f"The number of restaurants in this city is: {highest_restaurants_count}")
```

######  2. Calculate the average rating for restaurants in each city.


```python
#  Top 10 average rating restaurants with city

data['city with rating'] = data['City'] +' '+ data['Restaurant Name']+' '+data['Rating text']
data['city with rating'].value_counts().sort_values(ascending=False).head(10).to_frame()
```

###### 3. Determine the city with the highest average rating.


```python
rating_mapping = {
    'Excellent': 5,
    'Very Good': 4,
    'Good': 3,
    'Average': 2,
    'Poor': 1,
}

# Map the string ratings to numeric values using the mapping
df['Numeric_Rating'] = df['Rating text'].map(rating_mapping)

# Group the DataFrame by 'City' and calculate the mean numeric rating for each city
average_ratings_by_city = df.groupby('City')['Numeric_Rating'].mean()

# Get the city with the highest average rating
city_with_highest_avg_rating = average_ratings_by_city.idxmax()

# Get the highest average rating value
highest_avg_rating = average_ratings_by_city.max()

print(f"The city with the highest average rating is: {city_with_highest_avg_rating}",'\n')
print(f"The highest average rating is: {highest_avg_rating}")
```

#### Create a histogram or bar chart to visualize the distribution of price rangesamong the restaurants.

```python


# Assuming your data is in a pandas DataFrame called 'data'

# Define the broad price range categories
price_categories = ['Low', 'Medium', 'High']

# Group the 'Price range' column into the defined categories
df['Price Category'] = pd.cut(df['Price range'], bins=[0, 2, 3, float('inf')], labels=price_categories)

# Create a bar plot to visualize the count of restaurants in each price category
sns.barplot(data=df, x='Price Category',y='Restaurant ID')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

#### Calculate the percentage of restaurants in each price range category

```python
# see all unique values of price range
my_var =  df['Price Category'].value_counts()
print(my_var,'\n')

low = round(my_var['Low']/my_var.sum()*100,2)
print('Percentage of low range restaurents',low)

medium = round(my_var['Medium']/my_var.sum()*100,2)
print('Percentage of low medium restaurents',medium)

high = round(my_var['High']/my_var.sum()*100,2)
print('Percentage of low high restaurents',high)
```

```python
# Define the broad percentage categories
percentage_categories = ['low', 'medium', 'high']
values = [79.12, 14.74, 6.14]

# Create the pie chart
plt.pie(values, labels=percentage_categories, autopct='%1.1f%%')

# Add a title
plt.title('price range category')

# Display the chart
plt.show()
```
